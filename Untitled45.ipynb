```python
# %%
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
from google.colab import files
import time
import math
import warnings

# Suppress warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# %%
# Upload CSV files
print("Please upload your 2013.csv file:")
uploaded_2013 = files.upload()
print("\nPlease upload your 2014.csv file:")
uploaded_2014 = files.upload()

# Get file names
file_2013 = list(uploaded_2013.keys())[0]
file_2014 = list(uploaded_2014.keys())[0]

print(f"\nUsing files: {file_2013} and {file_2014}")

# %%
# Load datasets
def load_datasets(file_2013, file_2014):
    df_2013 = pd.read_csv(file_2013, skiprows=2)
    df_2014 = pd.read_csv(file_2014, skiprows=2)

    return df_2013, df_2014

# Load data
df_2013, df_2014 = load_datasets(file_2013, file_2014)

# %%
# Data Preprocessing
def preprocess_data(df_2013, df_2014):
    # Select relevant columns
    target_cols = ['Clearsky GHI', 'Clearsky DHI', 'Clearsky DNI', 'Wind Speed']

    # Add temperature to features if available
    all_cols = df_2013.columns.tolist()
    temp_col = [col for col in all_cols if 'temp' in col.lower() or 'temperature' in col.lower()]
    if temp_col:
        print(f"Found temperature column: {temp_col[0]}")
        target_cols.append(temp_col[0])

    # Find common columns between the datasets
    common_cols = list(set(df_2013.columns) & set(df_2014.columns))
    feature_cols = [col for col in common_cols if col not in target_cols + ['Year', 'Month', 'Day', 'Hour']]

    print(f"Using features: {feature_cols}")
    print(f"Targets: {target_cols}")

    # Create datetime index
    for df in [df_2013, df_2014]:
        df['Datetime'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour']])
        df.set_index('Datetime', inplace=True)

    # Extract features and targets
    X_2013 = df_2013[feature_cols].values
    y_2013 = df_2013[target_cols].values

    X_2014 = df_2014[feature_cols].values
    y_2014 = df_2014[target_cols].values

    # Normalize data
    x_scaler = StandardScaler()
    X_2013 = x_scaler.fit_transform(X_2013)
    X_2014 = x_scaler.transform(X_2014)

    y_scaler = StandardScaler()
    y_2013 = y_scaler.fit_transform(y_2013)

    return X_2013, y_2013, X_2014, y_2014, y_scaler, df_2014, feature_cols, target_cols

# Preprocess data
X_2013, y_2013, X_2014, y_2014, y_scaler, df_2014, feature_cols, target_cols = preprocess_data(df_2013, df_2014)

# ========================================
# Sequence Generator
# ========================================
def create_sequences(X, y, window_size=24):
    X_seq, y_seq = [], []
    for i in range(len(X) - window_size):
        X_seq.append(X[i:i+window_size])
        y_seq.append(y[i+window_size])
    return np.array(X_seq), np.array(y_seq)

# ========================================
# Autoformer Architecture
# ========================================
class SeriesDecomposition(nn.Module):
    def __init__(self, kernel_size):
        super().__init__()
        padding = (kernel_size - 1) // 2
        self.moving_avg = nn.AvgPool1d(kernel_size, stride=1, padding=padding, count_include_pad=False)

    def forward(self, x):
        x_perm = x.permute(0, 2, 1)
        trend = self.moving_avg(x_perm).permute(0, 2, 1)
        seasonal = x - trend
        return seasonal, trend

class DataEmbedding(nn.Module):
    def __init__(self, c_in, d_model, max_len=168):
        super().__init__()
        self.value_embedding = nn.Linear(c_in, d_model)
        self.position_embedding = nn.Parameter(torch.zeros(1, max_len, d_model))

    def forward(self, x):
        return self.value_embedding(x) + self.position_embedding[:, :x.size(1), :]

class AutoCorrelationLayer(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, queries, keys, values):
        scores = torch.matmul(queries, keys.transpose(-2, -1)) / math.sqrt(queries.size(-1))
        attn = torch.softmax(scores, dim=-1)
        return torch.matmul(attn, values)

class EncoderLayer(nn.Module):
    def __init__(self, d_model, kernel_size):
        super().__init__()
        self.decomp = SeriesDecomposition(kernel_size)
        self.attn = AutoCorrelationLayer()
        self.norm1 = nn.LayerNorm(d_model)
        self.ff = nn.Sequential(
            nn.Linear(d_model, d_model * 4),
            nn.ReLU(),
            nn.Linear(d_model * 4, d_model)
        )
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):
        seasonal, trend = self.decomp(x)
        x = self.norm1(x + self.attn(seasonal, seasonal, seasonal))
        x = self.norm2(x + self.ff(x))
        return x + trend

class DecoderLayer(nn.Module):
    def __init__(self, d_model, kernel_size):
        super().__init__()
        self.decomp = SeriesDecomposition(kernel_size)
        self.attn1 = AutoCorrelationLayer()
        self.attn2 = AutoCorrelationLayer()
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.ff = nn.Sequential(
            nn.Linear(d_model, d_model * 4),
            nn.ReLU(),
            nn.Linear(d_model * 4, d_model)
        )
        self.norm3 = nn.LayerNorm(d_model)

    def forward(self, x, enc_out):
        seasonal, trend = self.decomp(x)
        x = self.norm1(x + self.attn1(seasonal, seasonal, seasonal))
        x = self.norm2(x + self.attn2(x, enc_out, enc_out))
        x = self.norm3(x + self.ff(x))
        return x + trend

class Autoformer(nn.Module):
    def __init__(self, c_in, c_out, d_model, e_layers, d_layers, kernel_size):
        super().__init__()
        self.enc_emb = DataEmbedding(c_in, d_model)
        self.dec_emb = DataEmbedding(c_in + c_out, d_model)
        self.encoder_layers = nn.ModuleList([EncoderLayer(d_model, kernel_size) for _ in range(e_layers)])
        self.decoder_layers = nn.ModuleList([DecoderLayer(d_model, kernel_size) for _ in range(d_layers)])
        self.projection = nn.Linear(d_model, c_out)

    def forward(self, x_enc):
        enc = self.enc_emb(x_enc)
        for layer in self.encoder_layers:
            enc = layer(enc)

        last_step = x_enc[:, -1:, :]
        dec_input = torch.cat([last_step, torch.zeros(last_step.size(0), 1, self.projection.out_features).to(x_enc.device)], dim=-1)
        dec = self.dec_emb(dec_input)

        for layer in self.decoder_layers:
            dec = layer(dec, enc)

        return self.projection(dec[:, -1, :])

# ========================================
# Fixed Hyperparameters Configuration
# ========================================
fixed_config = {
    "window_size": 72,
    "d_model": 64,
    "e_layers": 3,
    "d_layers": 2,
    "batch_size": 32,
    "lr": 0.0005,
    "kernel_size": 73,
    "epochs": 200,
}

# === Train Final Model ===
def train_final_model(config):
    window = config["window_size"]
    X_seq, y_seq = create_sequences(X_2013, y_2013, window)

    # Create validation split (10%)
    split_idx = int(0.9 * len(X_seq))
    X_train, X_val = X_seq[:split_idx], X_seq[split_idx:]
    y_train, y_val = y_seq[:split_idx], y_seq[split_idx:]

    train_loader = DataLoader(
        TensorDataset(torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.float32)),
        batch_size=config["batch_size"],
        shuffle=True
    )

    val_loader = DataLoader(
        TensorDataset(torch.tensor(X_val, dtype=torch.float32), torch.tensor(y_val, dtype=torch.float32)),
        batch_size=config["batch_size"],
        shuffle=False
    )

    model = Autoformer(
        c_in=X_seq.shape[2],
        c_out=y_seq.shape[1],
        d_model=config["d_model"],
        e_layers=config["e_layers"],
        d_layers=config["d_layers"],
        kernel_size=config["kernel_size"]
    ).to(device)

    optimizer = optim.Adam(model.parameters(), lr=config["lr"])
    criterion = nn.MSELoss()

    # For tracking best model
    best_val_loss = float('inf')
    best_model_state = None
    train_losses = []
    val_losses = []

    for epoch in range(config["epochs"]):
        model.train()
        total_train_loss = 0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            loss = criterion(model(xb), yb)
            loss.backward()
            optimizer.step()
            total_train_loss += loss.item() * xb.size(0)

        avg_train_loss = total_train_loss / len(train_loader.dataset)
        train_losses.append(avg_train_loss)

        # Validation
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                val_loss = criterion(model(xb), yb)
                total_val_loss += val_loss.item() * xb.size(0)

        avg_val_loss = total_val_loss / len(val_loader.dataset)
        val_losses.append(avg_val_loss)

        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_model_state = model.state_dict().copy()

        print(f"Epoch {epoch+1}/{config['epochs']} | Train Loss: {avg_train_loss:.6f} | Val Loss: {avg_val_loss:.6f}")

    # Load best model state
    model.load_state_dict(best_model_state)
    return model, train_losses, val_losses

# ========================================
# Season Extraction Functions
# ========================================
def get_season_data(df, X, y, season_months):
    """Extract data for specific season months"""
    season_mask = df.index.month.isin(season_months)
    return df[season_mask], X[season_mask], y[season_mask]

# Define seasons
SUMMER_MONTHS = [3, 4]  # March, April
RAINY_MONTHS = [6, 7]   # June, July
WINTER_MONTHS = [12, 1, 2]  # December, January, February

# Extract season data
df_summer, X_summer, y_summer = get_season_data(df_2014, X_2014, y_2014, SUMMER_MONTHS)
df_rainy, X_rainy, y_rainy = get_season_data(df_2014, X_2014, y_2014, RAINY_MONTHS)
df_winter = df_2014[df_2014.index.month.isin(WINTER_MONTHS)]

# ========================================
# Season Evaluation Function
# ========================================
def evaluate_season(model, config, X_season, y_season, df_season, season_name, y_scaler):
    window = config["window_size"]
    # Create sequences (we need X sequences, but y_season is only used to determine length)
    X_season_seq, _ = create_sequences(X_season, np.zeros((len(X_season), y_season.shape[1])), window)
    X_season_t = torch.tensor(X_season_seq, dtype=torch.float32).to(device)

    model.eval()
    with torch.no_grad():
        y_pred_scaled = model(X_season_t).cpu().numpy()

    # Inverse scale predictions
    y_pred = y_scaler.inverse_transform(y_pred_scaled)
    actual = y_season[window:]

    # Metrics
    mse = mean_squared_error(actual, y_pred, multioutput='raw_values')
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual, y_pred, multioutput='raw_values')

    # Create metrics DataFrame
    metrics = pd.DataFrame({
        'Target': ['GHI', 'DHI', 'DNI', 'Wind Speed'],
        'MSE': mse[:4],
        'RMSE': rmse[:4],
        'MAE': mae[:4]
    })

    print(f"\nðŸ“Š {season_name} Season Evaluation Metrics (Modified Autoformer):")
    print(metrics)

    # Save metrics
    metrics.to_csv(f"{season_name}_season_metrics.csv", index=False)
    print(f"âœ… {season_name} season metrics saved")

    # Prepare results dataframe
    results_df = pd.DataFrame({
        'Datetime': df_season.index[window:],
        'Actual_GHI': actual[:, 0],
        'Predicted_GHI': y_pred[:, 0],
        'Actual_DHI': actual[:, 1],
        'Predicted_DHI': y_pred[:, 1],
        'Actual_DNI': actual[:, 2],
        'Predicted_DNI': y_pred[:, 2],
        'Actual_Wind_Speed': actual[:, 3],
        'Predicted_Wind_Speed': y_pred[:, 3]
    })

    # Add temperature if present
    if actual.shape[1] > 4:
        results_df['Actual_Temperature'] = actual[:, 4]
        results_df['Predicted_Temperature'] = y_pred[:, 4]

    # Save results
    results_df.to_csv(f"{season_name}_season_forecast_results.csv", index=False)
    print(f"âœ… {season_name} season results saved")

    return results_df, metrics

# ========================================
# Plotting Functions
# ========================================
def plot_season_results(results_df, season_name):
    fig, axes = plt.subplots(3, 1, figsize=(15, 18))
    fig.suptitle(f'ðŸ“ˆ {season_name} Season 2014 Forecast Results (Modified Autoformer)', fontsize=18)

    # GHI Plot
    axes[0].plot(results_df['Datetime'], results_df['Actual_GHI'], 'b-', label='Actual', linewidth=1.5)
    axes[0].plot(results_df['Datetime'], results_df['Predicted_GHI'], 'r--', label='Predicted', linewidth=1.2)
    axes[0].set_title('Global Horizontal Irradiance (GHI)', fontsize=14)
    axes[0].set_ylabel('GHI (W/mÂ²)', fontsize=12)
    axes[0].grid(True, alpha=0.3)
    axes[0].legend(fontsize=10)

    # DHI Plot
    axes[1].plot(results_df['Datetime'], results_df['Actual_DHI'], 'b-', label='Actual', linewidth=1.5)
    axes[1].plot(results_df['Datetime'], results_df['Predicted_DHI'], 'r--', label='Predicted', linewidth=1.2)
    axes[1].set_title('Diffuse Horizontal Irradiance (DHI)', fontsize=14)
    axes[1].set_ylabel('DHI (W/mÂ²)', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    axes[1].legend(fontsize=10)

    # DNI Plot
    axes[2].plot(results_df['Datetime'], results_df['Actual_DNI'], 'b-', label='Actual', linewidth=1.5)
    axes[2].plot(results_df['Datetime'], results_df['Predicted_DNI'], 'r--', label='Predicted', linewidth=1.2)
    axes[2].set_title('Direct Normal Irradiance (DNI)', fontsize=14)
    axes[2].set_ylabel('DNI (W/mÂ²)', fontsize=12)
    axes[2].set_xlabel('Date', fontsize=12)
    axes[2].grid(True, alpha=0.3)
    axes[2].legend(fontsize=10)

    # Format dates
    date_fmt = mdates.DateFormatter('%m-%d')
    for ax in axes:
        ax.xaxis.set_major_formatter(date_fmt)
        ax.tick_params(axis='x', rotation=45)

    plt.tight_layout(rect=[0, 0.03, 1, 0.96])
    plt.savefig(f"{season_name}_season_forecast_plot.png", dpi=300, bbox_inches='tight')
    plt.show()
    print(f"âœ… {season_name} season plot saved")

def plot_temperature_comparison(df_summer, df_winter):
    # Find temperature column
    temp_col = [col for col in df_summer.columns if 'temp' in col.lower() or 'Temperature' in col.lower()]

    if not temp_col:
        print("âš ï¸ Temperature column not found in dataset")
        return

    temp_col = temp_col[0]

    plt.figure(figsize=(15, 10))

    # Summer Temperature
    plt.subplot(2, 1, 1)
    plt.plot(df_summer.index, df_summer[temp_col], 'r-', linewidth=1.5)
    plt.title('Summer 2014 Actual Temperature (March-April)', fontsize=14)
    plt.ylabel('Temperature (Â°C)', fontsize=12)
    plt.grid(True, alpha=0.3)

    # Format dates for summer plot
    ax = plt.gca()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    plt.xticks(rotation=45)

    # Winter Temperature
    plt.subplot(2, 1, 2)
    plt.plot(df_winter.index, df_winter[temp_col], 'b-', linewidth=1.5)
    plt.title('Winter 2014 Actual Temperature (January-February & December)', fontsize=14)
    plt.ylabel('Temperature (Â°C)', fontsize=12)
    plt.xlabel('Date', fontsize=12)
    plt.grid(True, alpha=0.3)

    # Format dates for winter plot
    ax = plt.gca()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.savefig("temperature_comparison_2014.png", dpi=300, bbox_inches='tight')
    plt.show()
    print("âœ… Temperature comparison plot saved")

# ========================================
# Training History Plot
# ========================================
def plot_training_history(train_losses, val_losses):
    plt.figure(figsize=(12, 6))
    plt.plot(train_losses, label='Training Loss', linewidth=2)
    plt.plot(val_losses, label='Validation Loss', linewidth=2)
    plt.title('Model Training History', fontsize=16)
    plt.ylabel('Loss (MSE)', fontsize=12)
    plt.xlabel('Epoch', fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.savefig('training_history_modified_autoformer.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("âœ… Training history plot saved")

# ========================================
# Save Model
# ========================================
def save_model(model):
    torch.save(model.state_dict(), "modified_autoformer_forecast_model.pth")
    print("âœ… Model saved as: modified_autoformer_forecast_model.pth")

# ========================================
# Main Execution
# ========================================
if __name__ == "__main__":
    print("\n" + "="*50)
    print("ðŸš€ Starting Model Training with Fixed Hyperparameters")
    print("="*50)
    print(f"Window Size: {fixed_config['window_size']}")
    print(f"d_model: {fixed_config['d_model']}")
    print(f"Encoder Layers: {fixed_config['e_layers']}")
    print(f"Decoder Layers: {fixed_config['d_layers']}")
    print(f"Batch Size: {fixed_config['batch_size']}")
    print(f"Learning Rate: {fixed_config['lr']}")
    print(f"Kernel Size: {fixed_config['kernel_size']}")
    print(f"Epochs: {fixed_config['epochs']}")

    # Train model
    model, train_losses, val_losses = train_final_model(fixed_config)

    # Plot training history
    plot_training_history(train_losses, val_losses)

    # Evaluate and save results for SUMMER season
    print("\n" + "="*50)
    print("ðŸŒž Evaluating SUMMER Season (March-April 2014)")
    print("="*50)
    summer_results, summer_metrics = evaluate_season(
        model, fixed_config,
        X_summer, y_summer, df_summer,
        "Summer", y_scaler
    )
    plot_season_results(summer_results, "Summer")

    # Evaluate and save results for RAINY season
    print("\n" + "="*50)
    print("ðŸŒ§ï¸ Evaluating RAINY Season (June-July 2014)")
    print("="*50)
    rainy_results, rainy_metrics = evaluate_season(
        model, fixed_config,
        X_rainy, y_rainy, df_rainy,
        "Rainy", y_scaler
    )
    plot_season_results(rainy_results, "Rainy")

    # Plot temperature comparison
    plot_temperature_comparison(df_summer, df_winter)

    # Save model
    save_model(model)

    print("\n" + "="*50)
    print("âœ… All operations completed successfully!")
    print("="*50)
    print("ðŸ“‚ Generated Files:")
    print(f" - Summer_season_forecast_results.csv")
    print(f" - Rainy_season_forecast_results.csv")
    print(f" - Summer_season_metrics.csv")
    print(f" - Rainy_season_metrics.csv")
    print(f" - Summer_season_forecast_plot.png")
    print(f" - Rainy_season_forecast_plot.png")
    print(f" - temperature_comparison_2014.png")
    print(f" - training_history_modified_autoformer.png")
    print(f" - modified_autoformer_forecast_model.pth")
```
